# 8) Tensorboard

TensorBoard was utilized to monitor the training and evaluation phases, transforming log files into comprehensive visual dashboards. The platform provided real-time tracking of critical metrics, including the convergence of loss functions over 30,000 steps and the calculation of Mean Average Precision (mAP) during the evaluation of the test dataset. 

http://localhost:6006/

classification_loss measures how accurately the model identifies the class of an object (e.g., Oryx vs. Lion). The steady decline toward 0.05 shows high confidence in species identification.

localization_loss tracks the precision of the final bounding box coordinates. It measures the error between the predicted box and the ground-truth XML annotations to ensure the boxes tightly fit the detected animals. At the conclusion of 20,000 steps, the model achieved a localized raw loss as low as 0.03, with a stable smoothed average of 0.1. This indicates that the model has successfully learned to regress bounding boxes with high precision, significantly reducing the error from the initial training phases.

The RPN Localization Loss tracks how accurately the model's first-stage "proposals" find the location of an object. A lower value means the model is getting better at pinpointing where the animals are before it even tries to name them. By the end of training (20k steps), the smoothed value reached 0.04 with a raw value of 0.02, showing that the model has successfully converged and is predicting object locations with very high precision.

he RPN Objectness Loss evaluates the models ability to distinguish foreground objects from the background. By the end of 20k steps, this metric achieved a smoothed value of 0.03 and a raw value of 0.05. These results indicate that the model has successfully converged and can reliably identify areas containing objects with very high confidence.

The Normalized Total Loss shows a steady decline from the initial training phases, stabilizing significantly after the 10k step mark. This demonstrates a healthy training process where the model's global error was minimized

The Total Loss graph shows a successful and classic "learning curve". The trend began with a sharp decline, representing the initial phase where the model rapidly learned the most obvious features of the animals. By the 20,000-step mark, the curve has flattened (plateaued)
